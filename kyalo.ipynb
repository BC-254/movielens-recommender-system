```python
user_item = df.pivot_table(
    index = 'userId',
    columns = 'movieId',
    values = 'rating'
)
user_item
```

```python
user_item_filled = user_item.fillna(0)
user_item_filled
```

Computing User-User Similiarity using Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
user_sim = cosine_similarity(user_item_filled)
user_sim_df = pd.DataFrame(
    user_sim,
    index = user_item_filled.index,
    columns = user_item_filled.index
)
print(user_sim_df)
```

```python
def get_top_k_users(userId,k=5):
    sims = user_sim_df[userId].drop(userId)
    return sims.sort_values(ascending = False).head(k)

get_top_k_users(userId=1)
```

```python
def recommend_items(userId,k_users = 2,k_items=2):
    similar_users = get_top_k_users(userId,k_users)

    user_ratings = user_item.loc[userId]

    scores = {}

    for sim_user,similarity in similar_users.items():
        sim_user_ratings = user_item.loc[sim_user]

        for item, rating in sim_user_ratings.items():
            if pd.isna(user_ratings[item]) and not pd.isna(rating):
                scores[item] = scores.get(item, 0) + similarity * rating
    
    ranked_items = sorted(scores.items(), key = lambda x: x[1], reverse=True)

    movie_ids = [item for item, _ in ranked_items[:k_items]]
    titles = movies[movies['movieId'].isin(movie_ids)]['title'].tolist()
    return titles
```

```python
recommend_items(userId=1)
```
